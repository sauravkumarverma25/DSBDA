```python
arr=[20,56,89,90,22,45,34,66,78,987,12,88]
data=pd.DataFrame(arr)
data
```

```python
length=len(arr)
```

```python
length
```

```python
sum=0
for i in range(0,length):
    sum+=arr[i]
print("Sum: ",sum,"\n")
avg=sum/length
print("Mean: ",avg)
```

```python
num=0
for i in range(0,length):
    num+=(arr[i]-avg)**2
var=num/(length-1)
print("variance: ",var,"\n")
std=var**0.5
print("std: ",std)
```
